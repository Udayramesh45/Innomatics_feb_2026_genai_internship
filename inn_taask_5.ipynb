```python
# Program 1: Smart Parking Lot Management System (Modified Version)

def parking_lot_tracker(max_slots, entry_exit_records):
    vehicles_inside = 0
    highest_occupancy = 0
    for record in entry_exit_records:
        if record == "IN":
            vehicles_inside += 1
        elif record == "OUT" and vehicles_inside > 0:
            vehicles_inside -= 1
        highest_occupancy = max(highest_occupancy, vehicles_inside)
    if vehicles_inside > max_slots:
        parking_condition = "Capacity Exceeded"
    elif vehicles_inside == max_slots:
        parking_condition = "Full"
    else:
        parking_condition = "Available"
    print("Currently Parked Vehicles:", vehicles_inside)
    print("Peak Parking Usage:", highest_occupancy)
    print("Parking Status:", parking_condition)
total_capacity = 50
movement_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
parking_lot_tracker(total_capacity, movement_logs)
```

```python
# Problem 2: Online Food Delivery Time Estimator

def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5
    if traffic == "High":
        base_time += 10
    elif traffic == "Medium":
        base_time += 5
    elif traffic == "Low":
        base_time += 0
    if weather == "Rainy":
        base_time += 5
    elif weather == "Stormy":
        base_time += 10
    print("Estimated Delivery Time:", base_time, "minutes")
estimate_delivery_time(8, "High", "Rainy")
```

```python
# Program 3: Movie Theatre Seat Occupancy Analyzer

def analyze_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"
    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)
total_seats = 200
booked = [1] * 150
analyze_occupancy(total_seats, booked)
```

```python
# Problem 4: Cloud Server Load Classification System
def classify_server_load(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading
    average = total / len(cpu_readings)
    print("Average CPU Load:", round(average), "%")
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
classify_server_load([45, 60, 70, 85, 90])
```

```python
# Problem 5: Smart Classroom Resource Usage Monitor

def resource_usage_monitor(resource_usage):
    overused = []
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
resource_usage_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
# Problem 6: Online Event Registration Capacity Controller

def event_registration_controller(capacity, registrations):
    confirmed = 0
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
    waitlisted = registrations - confirmed
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if waitlisted > 0:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")
event_registration_controller(100, 105)
```
